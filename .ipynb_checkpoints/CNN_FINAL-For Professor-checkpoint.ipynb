```python
TRAINING_DIR=os.path.join(dataset_dir, '/train/')
TEST_DIR= os.path.join(dataset_dir, '/test/') #'/home/balahari/GTSRB_project/input/test/'
#Training Dataset Specifying paremeters for Image Data Generation
#Note: Only Training and Validation Image Dataset- Additional Constraints like Shear, Zoom are done and not on Test Set
training_datagen = ImageDataGenerator(
    rescale = 1./255,
    validation_split= .2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=False,
    fill_mode='nearest'
)
#Test Dataset Specifying paremeters for Image Data Generation(only normalised)
test_datagen = ImageDataGenerator(rescale = 1./255)

#This reads training data and splits 80% as TRAINING
#Input size of the image is 30x30, RGB mode
#Batch size is 32 by default, shuffle is TRUE by default
train_generator = training_datagen.flow_from_directory(
    TRAINING_DIR,
    target_size=(30,30),
    color_mode='rgb',
    class_mode='categorical',
    subset='training',
    #batch_size=32,
    #shuffle=TRUE,
    seed=24
)
#This reads training data and splits 80% as Validation
#Shuffle is TRUE by default, hence its overridden as False for validation data
validation_generator = training_datagen.flow_from_directory(
    TRAINING_DIR,
    target_size=(30,30),
    color_mode='rgb',
    class_mode='categorical',
    #batch_size=32,
    shuffle=False,
    subset='validation',
    seed=24
)
#This reads test data for prediction of models
#Shuffle is TRUE by default, hence its overridden as False for test data
test_generator = test_datagen.flow_from_directory(
    TEST_DIR,
    target_size=(30,30),
    color_mode='rgb',
    class_mode=None,
    batch_size=1,
    shuffle=False,
    seed=24
)

#Output should look like:
# Found 31368 images belonging to 43 classes.
# Found 7841 images belonging to 43 classes.
# Found 12630 images belonging to 1 classes.
```

# Creating the Model
1) Sequential model created with a total of 4 convolutions and 2 max pooling.(This provided better results than other configurations.) Activation Relu.

2) Model has 2 fully connected hidden layers, Activation Relu.

3) Output layer has 43 neurons with softmax activation.

4) 3 dropouts were tested with different values

Note: Models without dropouts performed better on training and validation data, but performed comparitively bad during prediction with test dataset.(Optimized for train data, but was not generalized)

Models with dropouts performed comparitively bad on training and validation data, but performed better on test dataset.(More Generalised Classifier was produced)

```python
model = tf.keras.models.Sequential([
    # The input shape is the desired size of the image 30x30 with 3 bytes color
    # This is the first convolution
    tf.keras.layers.Conv2D(32, (5,5), activation='relu', input_shape=(30, 30, 3)),
    # The second convolution
    tf.keras.layers.Conv2D(32, (5,5), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Dropout(0.2),
    # The third convolution
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    # The Fourth convolution
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Dropout(0.2),
    # Flatten the results to feed into a DNN
    tf.keras.layers.Flatten(),
    #  neuron hidden layer
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(43, activation='softmax')
])


model.summary()
```

## Training the model with data from train_generator and validation_generator

```python
#RMSprop optimizer is used and Learning rate for optimizer is set
opt = tf.keras.optimizers.RMSprop(learning_rate=0.001)

#Categorical crossentropy is preferred for this multi-class image classifier
model.compile(loss = 'categorical_crossentropy', optimizer=opt, metrics=['accuracy'])

#If Batch size is explicitly over ridden use the below values for steps_per_epoch and validation_steps
STEP_SIZE_TRAIN=train_generator.n//train_generator.batch_size
STEP_SIZE_VALID=validation_generator.n//validation_generator.batch_size

#Since Batch Size is not explicitly mentioned, use len(trian_generator) and len(validation_generator)
history = model.fit(train_generator, 
                    epochs=20, 
                    steps_per_epoch=len(train_generator), 
                    validation_data = validation_generator, 
                    verbose = 1, 
                    validation_steps=len(validation_generator)
                   )

```

## Calculating the accuracy of model on Training and Validation Dataset.

```python
#Graph to plot Training and validation accuracy & their loss
import matplotlib.pyplot as plt
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'r', label='Training accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')

plt.figure()

plt.plot(epochs, loss, 'r', label='Training Loss')
plt.plot(epochs, val_loss, 'b', label='Validation Loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()

#Printing Training and validation accuracy
print("Training accuracy", acc[-1])
print("Validation accuracy",val_acc[-1])
```
